```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load MLB team statistics CSV
teams = pd.read_csv('data/team_stats_2021.csv')
teams.head()
```

```python
# Drop columns with null values, they are not necessary
teams.dropna(axis = 1, inplace = True)
teams.head()
```

```python
teams.drop(['Unnamed: 2', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20', 
            'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26', 'Unnamed: 27', 
            'Unnamed: 28', 'Unnamed: 29', 'Unnamed: 30', 'Unnamed: 31', 'Unnamed: 32'], axis = 1, inplace = True)
teams.head()
```

```python
plt.figure(figsize = (10, 9))
ax = sns.scatterplot(x = 'W', y = 'Team', hue = 'OBP', data = teams)
plt.plot(95, 0, color = 'r')
```
